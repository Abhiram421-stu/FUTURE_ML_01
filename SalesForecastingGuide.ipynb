```python
import pandas as pd

url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/flights.csv"

df = pd.read_csv(url)

df.head()
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt

df['passengers'].plot()
plt.show()
df.head()
```

```python
df.head()
```

```python
df['date'] = pd.to_datetime(df['year'].astype(str) + '-' + df['month'])

df.head()
```

```python
df.set_index('date', inplace=True)

df.head()
```

```python
df['passengers'].plot(figsize=(10,5))
plt.title("Passengers Over Time")
plt.show()
```

```python
df['time'] = range(len(df))

df.head()
```

```python
from sklearn.linear_model import LinearRegression

X = df[['time']]
y = df['passengers']

model = LinearRegression()
model.fit(X, y)
```

```python
future = pd.DataFrame({'time': range(len(df), len(df)+12)})

predictions = model.predict(future)

predictions
```

```python
plt.plot(df['passengers'], label="Actual")
plt.plot(range(len(df), len(df)+12), predictions, label="Predicted", color='red')

plt.legend()
plt.title("Forecasting Future Passengers")
plt.show()
```

```python
future_dates = pd.date_range(start=df.index[-1], periods=13, freq='M')[1:]

plt.figure(figsize=(10,5))

plt.plot(df.index, df['passengers'], label="Actual")
plt.plot(future_dates, predictions, label="Predicted", color='red')

plt.legend()
plt.title("Forecasting Future Passengers")
plt.show()
```

# Sales/Demand Forecasting Project

## Objective
To predict future demand based on past data trends.

## Dataset
We used a time-series dataset containing:
- Year
- Month
- Number of passengers (demand)

## Steps Performed
1. Loaded dataset using pandas
2. Converted date into proper time format
3. Visualized trend using line graph
4. Created time-based feature
5. Applied Linear Regression model
6. Predicted future values (next 12 months)

## Result
The model shows an increasing trend in future demand.

## Limitation
The model does not capture seasonal variations, only overall trend.

## Conclusion
